### Select relevant columns needed

```python
df2_customized = df2[['ProductKey','Unit Cost USD','Unit Price USD']]
```

### Merging two datasets

```python
merged_df = customized_df.merge(df2_customized,on='ProductKey', how='inner')
```

```python
merged_df
```

### Removed the dollar sign
using slice
## Can also do this 
* cols_to_change = ['Unit Cost USD','Unit Price USD']

```python
merged_df['Unit Cost USD'] = merged_df['Unit Cost USD'].str.slice(1)
merged_df['Unit Price USD'] = merged_df['Unit Price USD'].str.slice(1)
```

```python
merged_df
```

```python
cols_to_change = ['Unit Cost USD','Unit Price USD']
merged_df[cols_to_change] = merged_df[cols_to_change].apply(lambda x: x.str.replace(',', '').str.strip().astype(float))
```

### Custom Column of Total Revenue per Order

```python
merged_df['Total Revenue per Order'] = (merged_df['Unit Price USD'] - merged_df['Unit Cost USD']) * merged_df['Quantity']
```

```python
merged_df
```

### Convert This to csv

```python
merged_df.to_csv('With_avg_total_sales.csv',index=False)
```

### Grouped and check the mean and sum based on purchase type

```python
grouped = merged_df.groupby('Purchase Type').agg({'Total Revenue per Order':['mean','sum']})
```
